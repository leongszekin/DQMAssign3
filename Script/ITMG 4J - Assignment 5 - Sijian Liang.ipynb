### 4.4 Convert Obesity Dataset's four Columns to Boolean Variables

Try to use the function I wrote in assignment 4 to clean the obesity dataset and convert "Family_History", "FAVC", "Smoking" and "SCC" to boolean values.

```python
def clean_column_types(df, num_cols, str_cols, bool_cols):
    """
    Ensure columns have the correct data types:
    - Convert numerical columns to numeric (coerce invalid values to NaN)
    - Convert string columns to string type
    - Convert boolean columns to boolean type
    """
     # Convert numerical columns
    df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')  # Force numeric conversion

    # Convert string columns
    df[str_cols] = df[str_cols].astype(str).apply(lambda x: x.str.strip())  # Strip spaces
    
   
    # Define the boolean conversion function
    def convert_to_bool(value):
        if pd.isna(value):  # Handle NaN cases
            return np.nan
        value = str(value).strip().lower()  # Normalize case and remove spaces
        if value in ['yes', '1', 'true', 'y', 't']:  # Accept different True formats
            return True
        elif value in ['no', '0', 'false', 'n', 'f']:  # Accept different False formats
            return False
        else:
            return np.nan  # If value is unknown, return NaN

    # Apply conversion function to boolean columns using apply() instead of map()
    for col in bool_cols:
        df[col] = df[col].astype(str).str.strip().apply(convert_to_bool)

    return df
```

```python
# Define column categories
num_cols = ["Age", "Height", "Weight", "FCVC", "NCP", "CH2O", 
            "FAF", "TUE"]
str_cols = ["Gender", "CAEC", "CALC","MTRANS","Obesity"]
bool_cols = ["Family_History", "FAVC", "Smoking","SCC"]

# Apply the function to clean column types
obesity_clean = clean_column_types(obesity, num_cols, str_cols, bool_cols)
```

## 5. Remove Duplicates BEFORE Merging

In step 1, I found that both of these two datasets have duplicate records per Age-Gender.

This would distort the analysis and misrepresent the relationship between heart attack risk and obesity level because duplicated records will inflate correlations and statistical significance incorrectly.

Therefore, removing duplicates before merging is important because it prevents the many-to-many explosion. 

But I didn't drop duplicates of any dataset because it may result in data loss and data integrity issues (If multiple people in the same Age-Gender group have different obesity levels, removing duplicates may lead to incorrect conclusions.)

To analyze the relationship between heart attack risk and obesity, the best way to do it is: "Aggregate each dataset to summarize key metrics so that each Age-Gender pair has one unique record. " 

Thus, I used "average" values to aggregate numeric columns and "most frequent value" to aggregate categorical.

Here are my steps:

1. Aggregate the Heart Attack Dataset to ensure each Age-Gender pair has one unique record
2. Check the rows left after aggregation of Heart Attack Dataset
3. Aggregate the Obesity Dataset to ensure each Age-Gender pair has one unique record
4. Check the rows left after aggregation of the Obesity Dataset
5. Use the aggregate results to merge to find out the relationship between heart attack risk and obesity




### 5.1 Aggregate the Heart Attack Dataset
Summarize key metrics so that each Age-Gender pair has one unique record.

The left join (how="left") is used in the merge operation to ensure that all Age-Gender pairs from the numeric summary (heart_attack_summary) are preserved, even if there are missing values in the categorical summary (heart_attack_modes).

```python
# Aggregating numeric features
heart_attack_summary = heart_attack.groupby(["Age", "Gender"]).agg({
    "BMI": "mean",  
    "Cholesterol_Level": "mean",  
    "Resting_BP": "mean",  
    "Heart_Rate": "mean", 
    "Max_Heart_Rate_Achieved": "mean"  
}).reset_index()

# Categorical columns to aggregate
categorical_cols = ["Smoking", "Alcohol_Consumption", "Diabetes", "Hypertension","Family_History", "Fasting_Blood_Sugar", "Exercise_Induced_Angina", 
                    "Physical_Activity_Level", "Stress_Level", "Chest_Pain_Type", 
                    "Thalassemia", "Fasting_Blood_Sugar", "ECG_Results", "Heart_Attack_Risk"]

# Aggregating categorical columns BEFORE merging with heart_attack_summary
heart_attack_modes = heart_attack.groupby(["Age", "Gender"])[categorical_cols].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None).reset_index()

# Merging numeric and categorical summaries
# Left join - the numeric summary of heart_attack_summary are more important, which should be preserved
heart_attack_summary = heart_attack_summary.merge(heart_attack_modes, on=["Age", "Gender"], how="left") 


# Display the aggregated dataset
heart_attack_summary.head(10)
```

### 5.3 Aggregate the Obesity Dataset
Similarly, aggregate relevant obesity data:

```python
obesity_summary = obesity_clean.groupby(["Age", "Gender"]).agg({
    # Aggregating numerical columns with mean or median
    "Height": "mean",  # Average height
    "Weight": "mean",  # Average weight
    "FCVC": "mean",  # Average frequency of vegetable consumption
    "NCP": "mean",  # Average number of main meals per day
    "CH2O": "mean",  # Average water intake (daily)
    "FAF": "mean",  # Average frequency of physical activity
    "TUE": "mean",  # Average screen time (daily)
}).reset_index()

# Aggregating categorical columns with mode (most frequent value)
categorical_cols = ["Family_History", "FAVC", "CAEC", "Smoking", "SCC", "CALC", "MTRANS", "Obesity"]

for col in categorical_cols:
    obesity_summary[col] = obesity_clean.groupby(["Age", "Gender"])[col].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None).reset_index(drop=True)

# Display the aggregated dataset
obesity_summary.head(10)
```

## 6. Merge datasets with common keys 

After ensuring "Age" and "Gender" are the same data types and removed duplicates, I merged two datasets on "Age" and "Gender" with "INNER JOIN".

#### - Why Use INNER JOIN?

__First, I only want patients present in both datasets to analyze the relationship between heart attack and obesity.__

An INNER JOIN keeps only the rows where there is a match in both datasets. 

- If someone appears in the heart attack dataset but not in the obesity dataset, I won't know their diet/exercise habits.
- If someone appears in the obesity dataset but not in the heart attack dataset, I won’t know their heart risk.

__Secondly, Keeping only complete records ensures better data quality. It avoids merging with missing data.__

- A LEFT JOIN would keep all heart attack records, even if no matching obesity data exists, leading to missing values in obesity-related columns.
- A RIGHT JOIN would keep all obesity records, but many wouldn’t have heart attack risk values.
- An INNER JOIN ensures every row has both heart risk and obesity data, reducing missing values.

```python
# Merge the datasets on Age and Gender
merged_df = heart_attack_summary.merge(obesity_summary, on=["Age","Gender"], how="inner")  
```
